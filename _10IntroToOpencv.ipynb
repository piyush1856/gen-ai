```python
import cv2
from google.colab.patches import cv2_imshow


# Load an image
image = cv2.imread('landscape.jpg')

# Display the image
cv2_imshow( image)
```

```python
# Resize the image
resized_image = cv2.resize(image, (500, 500))  # Specify new width and height

# Display the resized image
cv2_imshow(resized_image)
```

```python
# Crop the image
cropped_image = image[0:300, 0:300]  # Define the region to crop

# Display the cropped image
cv2_imshow(cropped_image)
```

```python
#Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
cv2_imshow(gray_image)
```

```python
# Adjust brightness and contrast
# alpha: contrast control (1.0-3.0), beta: brightness control (0-100)
alpha = 1.5  # Contrast control
beta = 50    # Brightness control

adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# Display the adjusted image
cv2_imshow(adjusted_image)
```

## SIFT detector
SIFT (Scale-Invariant Feature Transform) detector is a widely used algorithm in computer vision and image processing. It is designed to detect and describe local features in images that are invariant to image scale and rotation.

It's particularly useful for tasks like object recognition, image stitching

Lets see an overview of how it works
- **Find keypoints:** In each image it look for points that are different from their neighbors. These points are called keypoints. Keypoints are points that are distinctive and can be easily recognized.
- **Assign orientations:** Decide which way each keypoint is facing. Knowing the orientation of a keypoint helps to match it correctly, even if the image is rotated.
- **Create descriptors:** For each keypoint, create a description that tells the computer what the area around that point looks like. The descriptors capture the local image information around each keypoint, allowing SIFT to match keypoints that look similar. These are based on the same logic of edges, corners, colors etc.

For comparision or similarity checking:

- **Match keypoints:** Compare the descriptors of keypoints in one image with the descriptors of keypoints in another image. If the descriptors are similar, the keypoints probably correspond to the same object.

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Create a SIFT detector
sift = cv2.SIFT_create()

# Detect keypoints and descriptors
keypoints, descriptors = sift.detectAndCompute(gray_image, None)

# Draw keypoints
image_with_keypoints = cv2.drawKeypoints(image, keypoints, None)

# Display the keypoints
cv2_imshow(image_with_keypoints)
```
